```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn import neighbors, datasets
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/RBellaApriliaDamayanti22/Datasets/main/riceClassification.csv")
df
```

```python
# scaler = MinMaxScaler()
scaler = StandardScaler()
# scaler = RobustScaler()
```

```python
df_for_scaler = pd.DataFrame(df, columns = ['Area',	'MajorAxisLength',	'MinorAxisLength',	'Eccentricity',	'ConvexArea',	'EquivDiameter',	'Extent',	'Perimeter',	'Roundness',	'AspectRation'	])
df_for_scaler = scaler.fit_transform(df_for_scaler)
df_for_scaler = pd.DataFrame(df_for_scaler,columns = ['Area',	'MajorAxisLength',	'MinorAxisLength',	'Eccentricity',	'ConvexArea',	'EquivDiameter',	'Extent',	'Perimeter',	'Roundness',	'AspectRation'])
df_for_scaler
```

```python
df_drop_column_for_minmaxscaler=df.drop(['id', 'Area',	'MajorAxisLength',	'MinorAxisLength',	'Eccentricity',	'ConvexArea',	'EquivDiameter',	'Extent',	'Perimeter',	'Roundness',	'AspectRation'], axis=1)
df_drop_column_for_minmaxscaler
```

```python
df_new = pd.concat([df_for_scaler,df_drop_column_for_minmaxscaler], axis=1)
df_new.head()
```

```python
X=df_new.iloc[:,0:10].values
y=df_new.iloc[:,10].values
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,stratify=y, random_state=42)
```

```python
# estimators = [
#     ('rf_1', RandomForestClassifier(n_estimators=10, random_state=42)),
#     ('knn_1', KNeighborsClassifier(n_neighbors=10))             
# ]

# model = GaussianNB()
model = KNeighborsClassifier(n_neighbors=3)
# model = RandomForestClassifier(n_estimators = 100) 
# model = StackingClassifier(estimators=estimators, final_estimator=GaussianNB())
```

```python
model.fit(X_train, y_train)
Y_pred = model.predict(X_test) 

from sklearn import metrics
score=metrics.accuracy_score(y_test,Y_pred)
print(score)
```

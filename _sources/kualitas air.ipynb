```python
df = pd.read_csv("https://raw.githubusercontent.com/RBellaApriliaDamayanti22/Datasets/main/water_potability.csv")
df
```

```python
# scaler = MinMaxScaler()
scaler = StandardScaler()
# scaler = RobustScaler()
```

```python
df_for_scaler = pd.DataFrame(df, columns = ['ph',	'Hardness',	'Solids',	'Chloramines',	'Sulfate',	'Conductivity',	'Organic_carbon',	'Trihalomethanes',	'Turbidity'	])
df_for_scaler = scaler.fit_transform(df_for_scaler)
df_for_scaler = pd.DataFrame(df_for_scaler,columns = ['ph',	'Hardness',	'Solids',	'Chloramines',	'Sulfate',	'Conductivity',	'Organic_carbon',	'Trihalomethanes',	'Turbidity'])
df_for_scaler
```

```python
df_drop_column_for_minmaxscaler=df.drop(['ph',	'Hardness',	'Solids',	'Chloramines',	'Sulfate',	'Conductivity',	'Organic_carbon',	'Trihalomethanes',	'Turbidity'], axis=1)
df_drop_column_for_minmaxscaler
```

```python
df_new = pd.concat([df_for_scaler,df_drop_column_for_minmaxscaler], axis=1)
df_new.head()
```

```python
X=df_new.iloc[:,0:9].values
y=df_new.iloc[:,9].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,stratify=y, random_state=42)
```
